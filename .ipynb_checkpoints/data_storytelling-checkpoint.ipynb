```python
#import modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load cleaned data
liver_df = pd.read_csv('indian_liver_dataset_clean.csv', index_col=0)
print(liver_df.info())
liver_df.head()
```

```python
#convert sex to categorical
liver_df['sex'] = liver_df['sex'].astype('category')
liver_df.info()
```

```python
#view relationships between variables
sns.pairplot(liver_df, hue='liver disease', palette={True: 'red', False: 'blue'}, hue_order=[True, False])
plt.show()
```

### Sensitivity and specificity of blood tests

One notable observation from close inspection of the pair plot above is that in many cases patients that have liver disease have blood tests that are normal or close to normal. This indicates that these blood tests have low sensitivity. In other words, a normal result for these blood tests cannot rule out the presence of liver disease. 

On the other hand, it appears that some of the blood tests have very good specificity, or in other words a blood test result above a certain value virtually guarantees the presence of liver disease. 

These two conclusions, taken together, suggest that a machine learning algorithm trained on this data will have high specificity but low sensitivity. Any data point which falls far outside of the 'normal cluster' in the multidimensional space is almost guaranteed to be a patient with liver disease.

The high specificity and low sensitivity raise an issue with the potential value of this machine learning model; if the model cannot rule out liver disease, then it will be unable to give relevant input to a decision of whether or not to perform a liver biopsy on a patient. In cases where the model is confident that a patient has liver disease, a liver biopsy will still be required in order to diagnose the specific type of disease. In all other cases the model will be unable to make a strong prediction, and a liver biopsy will be required to rule out liver disease.

### Non-diseased outlier in ALP result

There is a striking sole outlier in the liver-disease-false (normal) group, in the ALP variable. This patient has an ALP value twice as high as any other patient in the normal group. It is on the high edge of the normal cluster on the bilirubin measurements, but is not strikingly different from the normal cluster in any other variable. This might be explained in a variety of ways:

- there was an error in the measurement of ALP
- there was an error in group assignment (i.e. this patient actually had liver disease, but was erroneously recorded as being disease-free)
- the patient had liver disease at the time of the blood test, but the disease spontaneously resolved by the time a liver biopsy was performed
- the patient is merely an outlier, and there is no error

It is unfortunately not possible, with this data alone, to determine which of the above possibilities is the truth.
